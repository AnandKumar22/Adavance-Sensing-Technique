```python
#declaring the array
a = np.zeros((28,28))
#print("Matrix A=",a)
#print(type(a))
b = np.zeros((28,1))
#print("Matrix B=",b)
#print(type(b))
for x in range(28):
    if((x%4)==0):
        b[x]=-5
    elif((x%4)==3):
        b[x]=+5
#print("Matrix B=",b)
#print(len(a))
#print(len(a[0]))
up = 4
for x in range(len(a)):
    for y in range(len(a[0])):
        if(y==x):
            a[x][y]=-4
        elif((y==(x+1)) and ((y%4)!=0)):
            a[x][y]=1
        elif((x==(y+1)) and ((x%4)!=0)):
            a[x][y]=1
for x in range(len(a)):
    for y in range(len(a[0])):
        if((y==(x+up)) or (x==(y+up))):
            a[x][y]=1
#print(a)
#print(b)
```

```python
#horizontal_distance and vertical_distance between grid 
hor_d = 0.162/5
ver_r = 6/8

#80 elements and 6 for P1,P2,P3,Q1,Q2,Q3
element_parameter = np.zeros((80,6))
for x in range(len(element_parameter)):
    if((x%2)==0):
        element_parameter[x][0]=0
        element_parameter[x][1]=ver_r
        element_parameter[x][2]=-ver_r
        element_parameter[x][3]=-hor_d
        element_parameter[x][4]=0
        element_parameter[x][5]=hor_d
    elif((x%2)!=0):
        element_parameter[x][0]=-ver_r
        element_parameter[x][1]=ver_r
        element_parameter[x][2]=0
        element_parameter[x][3]=0
        element_parameter[x][4]=-hor_d
        element_parameter[x][5]=hor_d
print(" ")
print("    P1      P2      P3      Q1      Q2      Q3")
print(element_parameter)

#Element Coefficient Matrix of Each Element
coeff = [None]*80
for x in range(len(coeff)):
    coeff[x] = np.zeros((3,3))
for x in range(len(coeff)):
    for y in range(len(coeff[x])):
        for z in range(len((coeff[x])[0])):
            (coeff[x])[y][z]=(((element_parameter[x][y]*element_parameter[x][z])+(element_parameter[x][y+3]*element_parameter[x][z+3]))/((element_parameter[x][1]*element_parameter[x][5])-(element_parameter[x][2]*element_parameter[x][4])))
print(" ")
print("Element Coefficient Matrix Of All Elements:")
print(" ")
print(coeff)

#Overall/Global Coefficient Matrix
#print("Overall/Global Coefficient Matrix:")
#overall = np.zeros((54,54))
#print(overall)
```

```python
volt = np.zeros((7,4))
for x in range(len(c)):
    volt[(int)(x/4)][(x%4)]=c[x]
pos_volt = 5*(np.ones((7,1)))
neg_volt = -5*(np.ones((7,1)))
#print(pos_volt,neg_volt)
#print(volt)
new_volt = np.hstack((pos_volt,volt,neg_volt))
print("Voltages at different nodes in solution region:")
print(" ")
print(new_volt)
print(" ")
d = 0.162/500
ele = np.zeros((7,5))
#print(ele)
for x in range(len(ele)):
    for y in range(len(ele[0])):
        ele[x][y] = (new_volt[x][y]-new_volt[x][y+1])/d
print("Electric Field at different region in solution region:")
print(" ")
print(ele)
```

```python
epsilon=8.854187871*10**-12
flux_density = 5*epsilon*ele
print("Flux Density:")
print(" ")
print(flux_density)
ref = np.ones((5,1))
net_flux_density = flux_density.dot(ref)
print(" ")
print("Net Flux Density in one direction:")
print(" ")
print(net_flux_density)
```

```python
length_of_plate = 0.06
width_of_plate = 0.01
Area_of_plate = length_of_plate*width_of_plate
Total_flux = net_flux_density[0][0]*Area_of_plate
print("Total Flux through Epoxy dielectic:",Total_flux)
```

```python
#From Gauss Law net flux passing through an area is equal to the charge enclosed within the gaussian surface
charge = Total_flux
Capacitance = charge/5
print("Capacitance is equal to =",Capacitance)
```
