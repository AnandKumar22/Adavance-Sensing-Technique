```python
#On differenttiating the energy matrix wrt vf and setting to zero. We get (Cff*vf = −Cfp*vp). 
#From this we find Vf(voltage) at free nodes.
#Cff(free to free node)
#Cff[i][j]=Overall Global Coefficient value of ith free node to jth free node
cff = np.zeros((1,14))
cff = [2, 3, 6, 7, 10, 11, 14, 15, 18, 19, 22, 23, 26, 27]
cff_matrix = np.zeros((len(cff),len(cff)))
for x in range(len(cff)):
    for y in range(len(cff)):
        cff_matrix[x][y]=overall[cff[x]-1][cff[y]-1]
for x in range(len(cff_matrix)):
    for y in range(len(cff_matrix[0])):
        cff_matrix[x][y]=round(cff_matrix[x][y], 10)
print("Cff_Matrix :")
print(" ")
print(cff_matrix)
```

```python
inverse_cff = slin.inv(cff_matrix)
print("Inverse Cff Matrix : ")
print(" ")
print(inverse_cff)
```

```python
#Vp (Voltage of prescribed Nodes)
Vp_matrix = np.zeros((14,1))
for x in range(len(Vp_matrix)):
    if(x%2==0):
        Vp_matrix[x][0]=5
    else:
        Vp_matrix[x][0]=-5
print("Vp Matrix :")
print(" ")
print(Vp_matrix)
```

```python
#Cfp matrix (Free to prescribed nodes)
cfp_row = cff
cfp_col = [1, 4, 5, 8, 9, 12, 13, 16, 17, 20, 21, 24, 25, 28]
cfp_matrix = np.zeros((len(cfp_row),len(cfp_col)))
for x in range(len(cfp_matrix)):
    for y in range(len(cfp_matrix[0])):
        cfp_matrix[x][y]=overall[cfp_row[x]-1][cfp_col[y]-1]
for x in range(len(cfp_matrix)):
    for y in range(len(cfp_matrix[0])):
        cfp_matrix[x][y]=round(cfp_matrix[x][y], 3)
print("Cfp Matrix :")
print(" ")
print(cfp_matrix)
```

```python
b_matrix = -1*((cfp_matrix).dot(Vp_matrix))
print("B_Matrix :")
print(" ")
print(b_matrix)
```

```python
#Vf = voltage at free nodes = (Cff)inverse * (B)
Vf_matrix = (inverse_cff).dot(b_matrix)
print("Voltages at free nodes :")
print(" ")
for x in range(len(Vf_matrix)):
    Vf_matrix[x][0]=round(Vf_matrix[x][0], 3)
    print("Voltage at free node",cff[x],": ",Vf_matrix[x][0])
#print(Vf_matrix)
```

```python
electric_field = np.zeros((7,1))
k=0
print("Electric Field in (V/m):")
print(" ")
for x in range(1, len(cff), 2):
    electric_field[k][0]=(Vf_matrix[x-1][0]-Vf_matrix[x][0])/(hor_d*10**-2)
    k+=1
print(electric_field)
```

```python
#According to Gauss Law,charge enclosed is equal to net flux crossing the surface area.
#Q(enclosed)=Electric Field * dielectric * ephsilon0 * Area
charge = electric_field[0][0]*5*8.854187817*10**-12*6*10**-4
print("Charge :",charge,"C")
```

```python
capacitance = charge/(5-(-5))
capacitance = capacitance*10**12
print("Capacitance :",capacitance,"pF")
```
