```python
import numpy as np
from numpy import *
import scipy.linalg as slin
```

```python
#declaring the array
np.set_printoptions(suppress=True)
diff_x = (0.162/5)
diff_y = (6/8)
diff_x_sq = diff_x**2
diff_y_sq = diff_y**2
a = np.zeros((28,28))
#print("Matrix A=",a)
#print(type(a))
b = np.zeros((28,1))
#print("Matrix B=",b)
#print(type(b))
for x in range(28):
    if((x%4)==0):
        b[x]=-5*(diff_y_sq)
    elif((x%4)==3):
        b[x]=+5*(diff_y_sq)
#print("Matrix B=",b)
#print(len(a))
#print(len(a[0]))
up = 4
for x in range(len(a)):
    for y in range(len(a[0])):
        if(y==x):
            a[x][y]=-2*(diff_x_sq + diff_y_sq)
        elif((y==(x+1)) and ((y%4)!=0)):
            a[x][y]=diff_y_sq
        elif((x==(y+1)) and ((x%4)!=0)):
            a[x][y]=diff_y_sq
for x in range(len(a)):
    for y in range(len(a[0])):
        if((y==(x+up)) or (x==(y+up))):
            a[x][y]=diff_x_sq
print(a)
print(b)
```

```python
print("Determinant of a=",slin.det(a))  #Determinant of the array
print("Trace of a=",np.trace(a))        #Trace of the array
#print("Transpose=",np.transpose(a))#Transpose of the array
#print("Inverse=",slin.inv(a))      #Inverse of the array
```

```python
#print((slin.inv(a)).dot(b))
#print("Length=",len((slin.inv(a)).dot(b)))
c = (slin.inv(a)).dot(b)
for x in range(len(c)):
    print("Calculated Voltage at Free Node ",x+1,"equals ",c[x])
```

```python
volt = np.zeros((7,4))
for x in range(len(c)):
    volt[(int)(x/4)][(x%4)]=c[x]
pos_volt = 5*(np.ones((7,1)))
neg_volt = -5*(np.ones((7,1)))
#print(pos_volt,neg_volt)
#print(volt)
new_volt = np.hstack((pos_volt,volt,neg_volt))
print("Voltages at different nodes in solution region:")
print(" ")
print(new_volt)
print(" ")
d = 0.162/500
ele = np.zeros((7,5))
#print(ele)
for x in range(len(ele)):
    for y in range(len(ele[0])):
        ele[x][y] = (new_volt[x][y]-new_volt[x][y+1])/d
print("Electric Field at different region in solution region:")
print(" ")
print(ele)
```

```python
epsilon=8.854187871*10**-12
flux_density = 5*epsilon*ele
print("Flux Density:")
print(" ")
print(flux_density)
ref = np.ones((5,1))
net_flux_density = flux_density.dot(ref)
print(" ")
print("Net Flux Density in one direction:")
print(" ")
print(net_flux_density)
```

```python
length_of_plate = 0.06
width_of_plate = 0.01
Area_of_plate = length_of_plate*width_of_plate
Total_flux = net_flux_density[0][0]*Area_of_plate
print("Total Flux through Epoxy dielectic:",Total_flux)
```

```python
#From Gauss Law net flux passing through an area is equal to the charge enclosed within the gaussian surface
charge = Total_flux
Capacitance = charge/5
print("Capacitance is equal to =",Capacitance)
```
